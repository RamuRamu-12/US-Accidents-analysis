## Start Time

```python
data.columns
```

```python
data.Start_Time[0]
```

```python
data.Start_Time=pd.to_datetime(data.Start_Time)
```

```python
data.Start_Time[0]
```

```python
## getting the hour at which the accident may done
accident_hour=data.Start_Time.dt.hour
```

```python
accident_hour
```

```python
#Plot describes the number of accidents done on each hour in a day.
sns.histplot(accident_hour,bins=24,stat="density")
```

```python
# Plot shows the  accidents on the particular day
sns.histplot(data.Start_Time.dt.day_of_week,bins=24)
```

```python
# To plot the accidnents on each hour on the particular day i.e Sunday.
sunday_start_time=data.Start_Time[data.Start_Time.dt.dayofweek==0]
sns.histplot(sunday_start_time.dt.hour,bins=24)
```

```python
# To plot the accidnents on each hour on the particular day i.e Friday.
friday_start_time=data.Start_Time[data.Start_Time.dt.dayofweek==5]
sns.histplot(friday_start_time.dt.hour,bins=24)
```

```python
#Accidents graph per month
sns.histplot(data.Start_Time.dt.month,bins=12)
```

```python
#Accidents graph per year
sns.histplot(data.Start_Time.dt.year,stat="percent")
```

```python
## Accidents done in a particular year month wise.
df_2019=data[data.Start_Time.dt.year==2019]
sns.histplot(df_2019.Start_Time.dt.month,bins=12)
```
